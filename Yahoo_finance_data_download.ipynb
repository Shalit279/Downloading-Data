```python
import yfinance as yf
import pandas as pd
import numpy as np

# Define ticker and dates
tickers = ["BRK-B", "^GSPC", "V", "MA", "WM", "JPM", "RY"]
start_date = "2015-12-01"
end_date = "2025-12-01"

# Download data
data = yf.download(tickers, start=start_date, end=end_date)

adjusted_close = data['Close']

daily_returns = np.log(adjusted_close / adjusted_close.shift(1)).dropna()

# Print the first few rows
print(daily_returns.to_string())

# Save to CSV
daily_returns.to_csv(f"{tickers}_daily_returns.csv")
```

```python
import yfinance as yf
import pandas as pd
import numpy as np

# --- Inputs ---
assets = ["^GSPC",  ]
market = "^GSPC"          # S&P 500 index in yfinance
start_date = "2015-12-01"
end_date = "2025-12-01"
TRADING_DAYS = 252

# --- Download adjusted close prices ---
prices = yf.download(assets, start=start_date, end=end_date, progress=False)["Close"]

# Drop rows where any price for any asset is missing
prices = prices.dropna()

# --- Compute daily returns (simple returns) ---
rets = prices.pct_change().dropna()

# Extract market returns
market_ret = rets[market]

# --- Volatility ---
daily_vol = rets.std()  # Daily volatility for all assets
annualized_vol = daily_vol * np.sqrt(TRADING_DAYS) # Annualized volatility for all assets

# --- Beta ---
# Calculate the full covariance matrix
cov_matrix = rets.cov()
# Extract the covariances of each asset with the market
cov_asset_market = cov_matrix[market]
# Calculate market variance
market_variance = market_ret.var()
# Calculate beta for all assets
beta = cov_asset_market / market_variance

print(f"Assets under consideration: {', '.join(assets)}")
print(f"Market index: {market}")
print("\nDaily Volatility:")
print(daily_vol.to_string(float_format='%.6f'))
print("\nAnnualized Volatility (252d):")
print(annualized_vol.to_string(float_format='%.6f'))
print(f"\nBeta vs {market}:")
print(beta.to_string(float_format='%.4f'))
```
